```python
import random

def generate_random_integers(n=1000, start=0, end=1000):
    random_integers = [random.randint(start, end) for _ in range(n)]
    return random_integers

random_integers = generate_random_integers()
random_integers = [str(i) for i in random_integers]
```

```python
def generate_random_floats(n=1000, start=10000, end=1000000):
    random_floats = [round(random.uniform(start, end), 2) for _ in range(n)]
    return random_floats

random_floats = generate_random_floats()
random_floats = [str(i) for i in random_floats]
```

```python
from trdg.generators import GeneratorFromStrings

from IPython.display import display, Image
from PIL import Image as PILImage
from io import BytesIO
from matplotlib.pyplot import imshow
import numpy as np

generator = GeneratorFromStrings(
    strings = random_floats,
    count = 300, # số ảnh sinh ra
    #fonts = ["/content/fonts/Gargi.ttf"],
    fonts = ["/content/drive/MyDrive/OCR/textgen/vi/SVN-Times New Roman Bold.ttf"],
    language="vi",
    size = 32,
    blur = 2.5,
    random_blur = True,
    background_type = 3,
    image_dir = '/content/drive/MyDrive/OCR/Dataset/background',
    text_color = '#000000,#000000',
    #distorsion_type=3
)
```
